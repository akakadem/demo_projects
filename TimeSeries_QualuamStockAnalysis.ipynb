```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing, SimpleExpSmoothing
```

```python
target = df_qualuam['Close']
```

```python
target
```

```python
len(target)
```

```python
# Naive Model -> is Previous Day Prices
target.shift(1)
```

```python
naive_historical_fit = target.shift(1)
```

```python
plt.figure(figsize=(12,6))
plt.plot(target.index, target, label='Actual Data')
plt.plot(target.index, naive_historical_fit, label='Naive Model')
plt.title('Apple Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.legend()
plt.grid('True')
plt.show
```

```python
ts_data = df_qualuam['Close']
len(ts_data)
```

```python
naive_historical_fit = ts_data.shift(1)
```

```python
plt.figure(figsize=(12,6))
plt.plot(ts_data.index, ts_data, label='Actual Data')
plt.plot(ts_data.index, naive_historical_fit, label='Naive Model')
plt.title('Apple Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.legend()
plt.grid('True')
plt.show
```

```python
#Naive Model: Before the we looked at 10 years of data, however, we will watch not less than 10 days
plt.figure(figsize=(12,6))
plt.plot(ts_data.index[-10:], ts_data[-10:], label='Actual Data')
plt.plot(ts_data.index[-10:], naive_historical_fit[-10:], label='Naive Model')
plt.title('Apple Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.legend()
plt.grid('True')
plt.show
```
